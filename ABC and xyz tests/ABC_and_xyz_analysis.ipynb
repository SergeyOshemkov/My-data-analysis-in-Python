##### Описание данных:

Продажи за год товаров бытовой химии в разбивке по месяцам.

Уже посчитана суммарная прибыль (profit) и количество проданных единиц (total) за год.

```python
import pandas as pd

import numpy as np
```

```python
data = pd.ExcelFile('./datasets/abc/abc_analysis.xlsx')
```

```python
df = data.parse('Data')

df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

#### Сортирую товары по суммарной прибыли от продаж:

```python
df_profit = df[['name', 'profit']].sort_values('profit', ascending=False)

df_profit.head()
```

#### Считаю вклад каждого из товаров в общую прибыль и кумулятивную прибыль:

```python
df_profit['contribution_share'] = df.profit \
                                    .div(df.profit.sum()) \
                                    .mul(100) \
                                    .round(1)

df_profit['cumulative_sum'] = df_profit.contribution_share.cumsum()

df_profit
```

 #### Буду считать группой "А" товары, приносящие 80 % прибыли, с 80% до 95% процентов - отнесу к группе "B" и остальные товары отнесу к группе "С".

```python
# Функция для разбивки товаров на классы:

def ABC(x):
    
    classs = ''
    
    if x < 81.0:
        classs = 'A'
    elif x >=81: 
        classs = np.where(x < 95.0, "B", "C")
        
    return classs
```

```python
df_profit['group'] = df_profit.cumulative_sum.apply(ABC)

df_profit
```
