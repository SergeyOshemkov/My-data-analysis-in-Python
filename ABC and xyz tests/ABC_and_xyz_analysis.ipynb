##### Описание данных:

Продажи за год товаров бытовой химии в разбивке по месяцам.

Уже посчитана суммарная прибыль (profit) и количество проданных единиц (total) за год.

```python
import pandas as pd

import numpy as np
```

```python
data = pd.ExcelFile('./datasets/abc/abc_analysis.xlsx')
```

```python
df = data.parse('Data')

df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df.name.unique()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

#### АВС анализ:

#### Сортирую товары по суммарной прибыли от продаж:

```python
df_profit = df[['name', 'profit']].sort_values('profit', ascending=False)

df_profit.head()
```

#### Считаю вклад каждого из товаров в общую прибыль и кумулятивную прибыль:

```python
df_profit['contribution_share'] = df.profit \
                                    .div(df.profit.sum()) \
                                    .mul(100) \
                                    .round(1)

df_profit['cumulative_sum'] = df_profit.contribution_share.cumsum()

df_profit
```

 #### Буду считать группой "А" товары, приносящие 80 % прибыли, с 80% до 95% процентов - отнесу к группе "B" и остальные товары отнесу к группе "С".

```python
# Функция для разбивки товаров на классы:

def ABC(x):
    
    classs = ''
    
    if x < 81.0:
        classs = 'A'
    elif x >=81: 
        classs = np.where(x < 95.0, "B", "C")
        
    return classs
```

```python
df_profit['profit_group'] = df_profit.cumulative_sum.apply(ABC)

df_profit
```

#### Провожу АВС-анализ по количеству проданных единиц:

```python
df_sold= df.drop(['profit'], axis=1).sort_values('total', ascending=False)

df_sold.head()
```

#### Нахожу вклад каждого товара в общее количество проданных единиц товара:

```python
df_sold['contribution_share'] = df_sold.total \
                                       .div(df_sold.total.sum()) \
                                       .mul(100) \
                                       .round(2)
```

#### Считаю кумулятивный вклад кадого товара по количеству проданных единиц товара:

```python
df_sold['cumulative_sum'] = df_sold.contribution_share \
                                     .cumsum() \
                                     .round(2)
df_sold
```

#### Разбиваю товары на группы ABC:

```python
df_sold['quantity_group'] = df_sold.cumulative_sum.apply(ABC)

df_sold
```

#### XYZ-анализ:

```python
df_xyz = df.drop(['profit', 'total'], axis=1)
```

#### Считаю среднее количество проданных штук каждого товара:

```python
df_xyz['mean_val'] = df_sold[['january',
                               'february',
                               'march',
                               'april',
                               'may',
                               'june',
                               'july',
                               'august',
                               'september',
                               'october',
                               'november',
                               'december']].mean(axis=1) \
                                           .round()
```

#### Считаю среднеквадратичное отклонение количества проданных штук каждого товара:

```python
df_xyz['stand'] = df_sold[['january',
                               'february',
                               'march',
                               'april',
                               'may',
                               'june',
                               'july',
                               'august',
                               'september',
                               'october',
                               'november',
                               'december']].std(axis=1) \
                                           .round()
```

#### Считаю вариацию по каждому товару чтобы понять насколько количество проданных единиц колеблется относительно среднего значения:

```python
df_xyz['variation'] = df_xyz.stand.div(df_xyz.mean_val) \
                            .mul(100) \
                            .round(2)
```

```python
df_xyz[['name', 'mean_val', 'stand', 'variation']].sort_values('variation', ascending=True)
```

#### Разобью на группы: при вариации 0-15% отнесу товары к группе х (самые стабильные продажи), при значениях вариации 15-25% - отнесу к группе y, при значениях вариации > 25% отнесу товары к группе z:

```python
# Функция, которая помогает разбить товары на группы в соответствии с выбранным правилом:

def xyz(x):
    
    classs = ''
    
    if x <= 15:
        classs = 'x'
    elif x >=15: 
        classs = np.where(x < 25, "y", "z")
        
    return classs
```

```python
df_xyz['xyz_group'] = df_xyz.variation.apply(xyz)
```

```python
df_xyz[['name', 'mean_val', 'variation', 'xyz_group']]
```

####  Объединяю в единую таблицу:

```python
df_analysis = df_profit[['name', 'profit_group']].merge(df_sold[['name', 'quantity_group']], on='name', how='inner') \
                                                 .merge(df_xyz[['name', 'xyz_group']],on='name', how='inner')

df_analysis
```

#### Выводы:

```python
df_analysis.query('profit_group == "A" & quantity_group == "A" & xyz_group == "x"')
```

#### Для этих товаров можно рекомендовать поддерживать минимальный прогнозируемый товарный запас, поскольку товар хорошо и регулярно продается.

```python
df_analysis.query('profit_group == "A" & quantity_group == "A" & xyz_group == "y"')
```

#### Поскольку эти товары приносят хорошую прибыль и продаются в большом количестве но не регулярно, можно рекомендовать деражать под рукой пул поставщиков, но не перетаривать ими склад чтобы избежать заморозки денег.

```python
df_analysis.query('profit_group == "A" & quantity_group == "A" & xyz_group == "z"')
```

#### Товары приносящие высокую прибыль и продающиеся в большом количестве, но абсолютно рандомно. Важно понять причину - возникли проблемы с поставкой или закончилась акция. Такой товар важно закупать в небольшом количестве впрок и хранить на складе на случай внезапного роста спроса чтобы избежать упущенной прибыли.

```python
df_analysis.query('profit_group == "B" & quantity_group == "B" & xyz_group == "z"')
```
